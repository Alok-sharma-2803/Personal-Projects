##### Question 1: Two Sum (Indices Version)
Return indices of two numbers that add up to target.

```python
nums = [2, 7, 11, 15]
target = 9

indice_1 = 0
indice_1 = 0
match_found = False

for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == target:
            indice_1 = i
            indice_2 = j
            match_found = True
            print(f"Following 2 indices hold the values whose sum matches {target}: {indice_1}, {indice_2}")
if match_found == False:
    print(f"No 2 values from the list {nums} sum up to {target}")
```

##### Question 2: Group Anagrams
Group words that are anagrams.

```python
words = ["eat", "tea", "tan", "ate", "nat", "bat"]

def sort_string(x_str):
    return ''.join(sorted(x_str))

temp_dict = {}

for i in words:
    print(i)
    if sort_string(i) in temp_dict.keys():
        temp_dict[sort_string(i)].append(i)
    else:
        temp_dict[sort_string(i)] = []
        temp_dict[sort_string(i)].append(i)

list(temp_dict.values())
```

##### Question 4: Find Missing Number
Find the missing number from 1 to n.

```python
nums = [1, 2, 4, 5, 6]
nums_min = min(nums)
nums_max = max(nums)

for i in range(nums_min, nums_max + 1):
    if i not in nums:
        print(i)
```

##### Question 6: Longest Word in List
Return longest word.
If tie → return first occurring.

```python
words = ["apple", "banana", "kiwi", "strawberry"]

word = ""

for i in words:
    if len(i) > len(word):
        word = i
print(word)
```

##### Question 7: Flatten One-Level Nested List
Only 1 level deep.

```python
nested = [[1,2], [3,4], [5]]
flattened_list = []

for i in nested:
    for j in i:
        flattened_list.append(j)
print(flattened_list)
```

##### Question 9: Find First Non-Repeating Character

```python
s = "aabbccdeff"

temp_dict = {}
for i in s:
    if i in temp_dict.keys():
        temp_dict[i] += 1
    else:
        temp_dict[i] = 1

non_repeating_chars = []
for key, value in temp_dict.items():
    if value == 1:
        non_repeating_chars.append(key)

print(non_repeating_chars[0])

```

##### Question 11: Merge Two Dictionaries (Sum Common Keys)

```python
d1 = {"a": 10, "b": 20}
d2 = {"b": 5, "c": 15}
output_dict = d1.copy()

for i in d2.keys():
    if i in output_dict.keys():
        output_dict[i] += d2[i]
    else:
        output_dict[i] = d2[i]

output_dict
```

##### Question 13: Check if List is Palindrome (No Reverse Built-in)

```python
nums = [1,2,3,2,1]

nums_rev = [nums[x] for x in range(len(nums) - 1, -1, -1)]
if nums == nums_rev:
    print(f"{nums} list is Palindrome")
else:
    print(f"{nums} list is not a Palindrome")
```

##### Question 14: Find All Pairs with Given Difference
Order doesn’t matter.

```python
nums = [1,5,3,4,2]
diff = 2

nums_unique = list(set(nums))
list_pairs = []
for i in range(len(nums_unique)):
    for j in range(i + 1, len(nums_unique)):
        if nums_unique[i] - nums_unique[j] == diff or nums_unique[j] - nums_unique[i] == diff:
            pair_1 = tuple([nums_unique[i], nums_unique[j]])
            pair_2 = tuple([nums_unique[j], nums_unique[i]])
            if pair_1 not in list_pairs and pair_2 not in list_pairs:
                list_pairs.append(pair_1)
list_pairs
```

##### Question 15: Top K Frequent Elements
Don’t just sort full dictionary blindly if avoidable.

```python
nums = [1,1,1,2,2,3]
k = 2

temp_dict = {}
for i in nums:
    if i in temp_dict.keys():
        temp_dict[i] += 1
    else:
        temp_dict[i] = 1

temp_dict
max_keys = []

if len(list(temp_dict.keys())) >= k:
    for i in range(k):
        max_keys.append(max(temp_dict, key = temp_dict.get))
        temp_dict.pop(max(temp_dict, key = temp_dict.get))
else:
    for i in range(len(temp_dict.keys())):
        max_keys.append(max(temp_dict, key = temp_dict.get))
        temp_dict.pop(max(temp_dict, key = temp_dict.get))
max_keys
```
