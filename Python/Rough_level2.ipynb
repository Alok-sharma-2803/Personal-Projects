##### Question 1
Print a Right-Angled Triangle Pattern and it's inverted version

```python
# Answer 1
n = 5
# Right-angled triangle
for i in range(1, n + 1):
    print("*" * i)
print('\n')

# Inverted right-angled triangle
for i in range(n, 0, -1):
    print('*' * i)
```

##### Question 2
Print a Pyramid Pattern

```python
# Answer 2

n = 5
var_int = 1
for i in range(1, n + 1):
    star = '*' * var_int
    space = ' ' * (n - i)
    print(f"{space}{star}{space}")
    var_int += 2
```

##### Question 3
Write a function that:

Takes a number

Returns whether it is prime

Then:

Print all prime numbers between 1 and 100

```python
# Answer 3

def is_prime(num):
    is_prime = "is a prime"
    if num == 2:
        is_prime = "is neither a prime nor a composite"
    else:
        for i in range(2, num):
            if num % i == 0:
                is_prime = "is not a prime"
    return is_prime

try:
    user_input = int(input("Please enter a number: "))
    print(f"You have entered {user_input}")
    print(f"{user_input} {is_prime(user_input)} number")
except ValueError:
    print("Invalid entry! A number was expected")
```

##### Question 4
Print first n Fibonacci numbers.

Then:

- Store them in a list

- Print their sum

```python
# Answer 4

def fibonacci_numbers(n):
    num1 = 0
    num2 = 1
    fib_list = []
    for i in range(n):
        if i < 2:
            fib_list.append(i)
        else:
            temp = fib_list[i - 2] + fib_list[i - 1]
            fib_list.append(temp)
    return fib_list

try:
    user_input = int(input("Please enter a number (n): "))
    print("You have entered {user_input}")
    var_list = fibonacci_numbers(user_input)
    print(f"The fibonacci series (up to n digits) is: {var_list}\nThe sum of this series is {sum(var_list)}")
except ValueError:
    print("Invalid entry! A number was expected")
```

##### Question 6
Find Frequency of Each Element
nums = [1,2,3,2,4,1,3,1,2]
Output:
- 1 appears 3 times
- 2 appears 3 times
- 3 appears 2 times
- 4 appears 1 time

```python
# Answer 6

nums = [1,2,3,2,4,1,3,1,2]
unique_values = []

for i in nums:
    if i not in unique_values:
        unique_values.append(i)

for i in unique_values:
    count = 0
    for j in nums:
        if j == i:
            count += 1
    print(f"{i} appears {count} times") 
```

##### Question 9
Check if a string is palindrome. Ignore case and spaces

```python
# Answer 9

var_str = "Race car"

var_str_treated = var_str.replace(' ', '').lower()
var_str_palindrome = ''
for i in range(len(var_str_treated) - 1, -1, -1):
    var_str_palindrome += var_str_treated[i]

if var_str_treated == var_str_palindrome:
    print(f"{var_str} is a palindrome")
else:
    print(f"{var_str} is not a palindrome")
```
