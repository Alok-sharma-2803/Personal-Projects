##### Question 1: Sum of Even Numbers

Given a list of integers, return the sum of only the even numbers.

```python
nums = [4, 7, 10, 3, 8, 5]
sum_of_odds = sum([x for x in nums if x % 2 == 0])
print(sum_of_odds)
```

##### Question 2: Remove Duplicates (Preserve Order)

Given a list, remove duplicates but keep the original order.

```python
nums = [1, 2, 2, 3, 4, 3, 5]
nums_unique = []
for i in nums:
    if i not in nums_unique:
        nums_unique.append(i)
print(nums_unique)
```

##### Question 3: Find the Second Largest Number

```python
nums = [10, 5, 8, 20, 15]
nums_copy = nums.copy()
nums_copy.sort()
# nums_sorted = nums.copy()
# nums_sorted.sort(reverse = True)
# nums_sorted[1]

x_largest = 2

for i in range(x_largest - 1):
    nums_copy.pop()

print(max(nums_copy))
```

##### Question 6: Word Frequency Counter

Given a list of words, create a dictionary that counts frequency of each word.

```python
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
output_dict = {}

for i in words:
    if i in output_dict.keys():
        output_dict[i] += 1
    else:
        output_dict[i] = 1

output_dict
```

##### Question 7: Find Student with Highest Marks
Return the name of the student with highest marks.

```python
marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78
}
marks_list = list(marks.values())

for i in marks.keys():
    if marks[i] == max(marks_list):
        print(i)
```
