##### Question 2

Take an integer input from the user.
Print whether it is: Positive, Negative or Zero

```python
# Answer 2
def num_type(x):
    if x > 0:
        print(f"Number {x} is Positive")
    elif x < 0:
        print(f"Number {x} is Negative")
    else:
        print(f"Number {x} is Zero")

try:
    user_input = int(input("Please enter an integer: "))
    print(f"You entered {user_input}")
    num_type(user_input)
except ValueError:
    print("Invalid input! Integer was asked.")
```

##### Question 5
Given a number:
1. If divisible by 3 → print "Fizz"
2. If divisible by 5 → print "Buzz"
3. If divisible by both → print "FizzBuzz"

```python
# Answer 5

def fizz_buzz_fizzBuzz(num):
    if num % 3 == 0 & num % 5 == 0:
        return "FizzBuzz"
    elif num % 3 == 0:
        return "Fizz"
    elif num % 5 == 0:
        return "Buzz"
    else:
        pass

try:
    user_input = int(input("Please enter a number: "))
    print(f"You have entered: {user_input}\nOutput = {fizz_buzz_fizzBuzz(user_input)}")
except ValueError:
    print("Invalid entry! A number was expected")
```

##### Question 6
Given a score (0–100), print:
1. A (90+)
2. B (75–89)
3. C (50–74)
4. F (<50)

```python
# Answer 6

def score_grading(num):
    if num < 50:
        return "F"
    elif num < 75:
        return "C"
    elif num < 90:
        return "B"
    elif num <= 100:
        return "A"
    else:
        pass

try:
    user_input = float(input("Please enter a score between 0-100: "))
    if (user_input > 100 or user_input < 0):
        print("Invalid entry! A number between 0 and 100 was expected.")
    else:
        print(f"You have entered a score of {user_input}. The grade for this score is {score_grading(user_input)}")
except ValueError:
    print("Invalid entry! A number was expected as score")
```

##### Question 14
reverse a list w/o using the built-in reverse function
